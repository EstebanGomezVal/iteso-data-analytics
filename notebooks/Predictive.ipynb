```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_features = 1, noise = 200)
```

```python
X.shape, y.shape
```

```python
import pandas as pd

df = pd.DataFrame(X)
df.columns = ['x']
df['y'] = y
df
```

```python
df['x'].plot(kind ='hist')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 5))
plt.scatter(X, y, label = 'Random Data', c = 'red')
plt.plot(X, y, label = "Plot", c = 'green')

plt.xlabel("Features")
plt.ylabel("Target (y)")
plt.title('Scatter Example plot')
plt.legend()
```

```python
import numpy as np

y_mean = np.array([y.mean()]*100)
```

```python
from sklearn.linear_model import LinearRegression

model_1 = LinearRegression()
model_2 = LinearRegression()
model_3 = LinearRegression()

model_1.fit(X, y)
```

```python
y_prend = model_1.predict(X)
y_prend
```

```python
X, y = make_regression(n_features = 1, noise = 30, random_state= 0)
plt.figure(figsize = (10, 4))

plt.scatter(X, y, label = 'Random Data', c = 'black')

plt.plot(X, y_mean, label = "Mean", c = 'green')

plt.plot(X, y_prend, label = "Linear Regression", c = 'blue')

plt.xlabel("Features")
plt.ylabel("Target (y)")
plt.title('Scatter Example plot')
plt.legend()

plt.savefig('../reports/figures/random1.svg')
```
